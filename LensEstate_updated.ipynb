</h1>

<p> The real estate sector represents one of the most economically and socially significant markets, directly influencing investment decisions, urban development, and household wealth. In emerging markets, however, real estate ecosystems often lack transparency, standardized information, and analytical infrastructure. 
    With the rapid expansion of digital platforms and social media, large volumes of property-related data are generated daily, yet much of this information remains unstructured and underutilized. <h2 style="color: #7b98a6;" > 🛠️ 1 Market Context :</h2> 
<p>The real estate market in Tunisia operates within a fragmented and partially informal ecosystem where property information is dispersed across multiple digital platforms, social media channels, and offline networks. Listings are often inconsistent, incomplete, or unstructured, making it difficult to obtain a unified and reliable view of the market. This lack of standardization limits transparency and complicates comparative analysis across regions and property types.</p>
<h2 style="color: #7b98a6;" > 🛠️  2 Problem Statement :</h2>
<p>Despite the growing availability of online listings, the market still lacks structured analytical mechanisms capable of transforming raw property data into actionable insights. Pricing decisions are frequently subjective, valuation practices vary significantly, and regional market dynamics remain insufficiently understood. The absence of predictive tools and intelligent systems prevents stakeholders from making fully informed investment, pricing, and planning decisions.</p>
<h2 style="color: #7b98a6;" > 🛠️ 3 Project Motivation :</h2>
<p>As digital transformation accelerates across industries, the real estate sector presents a strong opportunity for intelligent automation and data-driven innovation. Addressing these needs requires more than simple data aggregation — it calls for advanced analytical capabilities that can interpret, predict, and explain market behavior.</p>

<h2 style="color: #426d84;"> 📌 1 — Methodology: Team Data Science Process (TDSP)</h2>

<h2 style="color: #426d84;"> 🏗️ 3 — System Architecture (Detailed Overview)</h2>

<p>
The proposed architecture is designed as a modular, agent-based, and scalable AI ecosystem structured around two major layers: the <b>Data Layer</b> and the <b>Intelligence & Analytics Layer</b>. The architecture follows a pipeline-oriented yet feedback-enabled design, allowing continuous learning, data enrichment, and model refinement.
</p>

<p>
Each component of the architecture corresponds directly to one or more Data Science Objectives (DSOs), ensuring that business strategy is tightly aligned with technical implementation.
</p>



<h3 style="color:#5c8399;"> 🔹 1. Data Layer — Intelligent Data Infrastructure</h3>

<p>
The Data Layer is responsible for transforming fragmented, heterogeneous, and partially unstructured real estate information into a standardized, high-quality analytical dataset. It operates through a coordinated multi-agent system supervised by a <b>Master Orchestrator Agent</b>.
</p>

<h4>🔸 Master Orchestrator Agent</h4>
<p>
This agent acts as the central coordinator of the data ecosystem. It schedules scraping tasks, monitors data quality checks, triggers cleaning workflows, and ensures synchronization between data sources. It also handles logging, version control of datasets, and error management.
</p>

<h4>🔸 Listing Scraper Agent</h4>
<p>
This autonomous agent collects property listings from digital platforms (e.g., Tayara, Mubawab) and potentially social media channels. It extracts:
</p>
<ul>
<li>Structured attributes (price, location, surface, property type)</li>
<li>Unstructured textual descriptions</li>
<li>Property images</li>
<li>Timestamps and listing metadata</li>
</ul>

<p>
The agent is designed to handle pagination, duplicate detection, anti-bot constraints, and dynamic content extraction.
</p>

<h4>🔸 Economic & Regional Data Agent</h4>
<p>
This agent retrieves macroeconomic, demographic, and infrastructural indicators such as:
</p>
<ul>
<li>Regional income levels</li>
<li>Population density</li>
<li>Interest rates</li>
<li>Construction activity</li>
<li>Urban development signals</li>
</ul>

<p>
These features are critical for DSO3 (regional trend forecasting) and enhance valuation context modeling.
</p>

<h4>🔸 Data Cleaning & Standardization Agent</h4>
<p>
Once raw data is collected, this agent performs:
</p>
<ul>
<li>Missing value handling</li>
<li>Schema normalization</li>
<li>Currency and unit standardization</li>
<li>Deduplication across platforms</li>
<li>Outlier detection</li>
</ul>

<p>
The output is a centralized structured database that feeds the Intelligence Layer.
</p>



<h3 style="color:#5c8399;"> 🔹 2. Intelligence & Analytics Layer — AI Decision Core</h3>

<p>
This layer transforms structured data into actionable intelligence. It consists of specialized analytical modules, each corresponding to a strategic capability.
</p>



<h4>🔸 Multi-Modal Valuation Engine (DSO2)</h4>

<p>
This module integrates three complementary modeling pipelines:
</p>

<ul>
<li><b>Structured Modeling Pipeline:</b> Gradient boosting / regression models trained on numerical and categorical attributes.</li>
<li><b>NLP Pipeline:</b> Language models analyze property descriptions to detect qualitative signals such as luxury level, renovation status, urgency bias, or promotional language.</li>
<li><b>Computer Vision Pipeline:</b> CNN/ViT-based models extract visual signals from property images (condition, brightness, space perception, finishing quality).</li>
</ul>

<p>
A fusion layer combines outputs from these pipelines into a unified valuation score. Explainability mechanisms (feature importance, SHAP values) ensure interpretability.
</p>



<h4>🔸 Regional Trend Forecasting Agent (DSO3)</h4>

<p>
This module models price evolution at multiple geographic levels (city, district, neighborhood). It integrates:
</p>

<ul>
<li>Historical transaction data</li>
<li>Macroeconomic indicators</li>
<li>Demand-supply signals</li>
<li>Urban expansion patterns</li>
</ul>

<p>
Advanced time-series and machine learning models generate forward-looking indicators such as:
</p>
<ul>
<li>Price growth projections</li>
<li>Volatility indexes</li>
<li>Emerging zone detection</li>
<li>Risk exposure scores</li>
</ul>

<p>
This agent supports strategic investment and planning decisions.
</p>



<h4>🔸 Conversational AI Module (DSO4)</h4>

<p>
This module connects structured data and analytical outputs to a large language model (LLM) via Retrieval-Augmented Generation (RAG). It enables users to:
</p>

<ul>
<li>Ask natural language queries</li>
<li>Request property comparisons</li>
<li>Generate analytical summaries</li>
<li>Receive investment recommendations</li>
</ul>

<p>
The conversational agent does not generate generic responses; instead, it grounds its answers in structured market data retrieved dynamically from the database.
</p>



<h4>🔸 Spatial Reconstruction & Visualization Engine (DSO5)</h4>

<p>
This module enhances spatial interpretability by generating 2D and 3D representations of properties. It combines:
</p>

<ul>
<li>Layout inference models</li>
<li>Vision-based structural feature extraction</li>
<li>Geometric consistency validation</li>
</ul>

<p>
The goal is not architectural precision but spatial coherence, allowing users to better understand volume, layout distribution, and structural organization.
</p>



<h3 style="color:#5c8399;"> 🔹 3. Integration & Data Flow</h3>

<p>
Data flows from the Data Layer into a centralized structured repository. The Intelligence Layer consumes this repository through API-based or pipeline-based access. Each analytical module produces enriched outputs (valuation scores, forecasts, explanations, 3D models) that are reintegrated into the system and exposed through user interfaces or dashboards.
</p>

<p>
The architecture supports:
</p>

<ul>
<li>Scalability (new agents can be added)</li>
<li>Modularity (independent model updates)</li>
<li>Continuous retraining</li>
<li>Future integration with external APIs</li>
</ul>



<h3 style="color:#5c8399;"> 🔹 Architectural Philosophy</h3>

<p>
The system follows an agent-based intelligence paradigm where autonomous components specialize in specific analytical functions while remaining orchestrated within a unified ecosystem. This design ensures robustness, extensibility, and long-term adaptability to evolving market conditions.
</p>

<p>
Rather than functioning as a simple listing aggregator, the architecture operates as a full-stack real estate intelligence platform capable of transforming raw digital signals into structured, predictive, and explainable market knowledge.
</p>

<p style="font-size:15px;">
Data standardization is the process of transforming raw, heterogeneous datasets collected from multiple platforms into a single, schema-consistent, analysis-ready master dataset. Each real estate platform in this pipeline (Mubawab, Tayara, BigDatis, and Tunisie Annonce) uses different column names, different languages (French vs. English), different categorical label conventions, and different levels of field granularity. <h2 style="color: #87BAC3;">📥 Loading Mubawab Listings</h2>

<h3>🎯 Purpose</h3>
<p>
This first step focuses on loading all raw CSV files collected from the <strong>Mubawab</strong> platform into a single unified DataFrame. Since the scraping process was split into multiple runs — each producing its own CSV file — we need to aggregate them into one cohesive dataset before any processing can begin.
</p>

<h3>🔍 Technical Breakdown</h3>
<p>
The <code>glob.glob()</code> function scans the <code>Data/raw_listings/mubaweb/</code> directory and returns the paths of all <code>.csv</code> files stored there. Each file corresponds to a scraping batch. These files are then read individually using <code>pd.read_csv()</code> inside a generator expression and concatenated using <code>pd.concat()</code> with <code>ignore_index=True</code>, which resets the row index to avoid conflicts between files. The result is a single flat DataFrame called <code>mubaweb_df</code> containing all Mubawab listings discovered across all scraping runs. The <code>.head()</code> call at the end gives us a quick preview of the first rows to visually verify the data was loaded correctly.
</p>

<h3>⚠️ Why This Matters</h3>
<p>
Keeping data in separate files is a common practice when scraping in batches, but it creates fragmentation. Merging all files at this early stage ensures a single consistent working object throughout the pipeline. Failing to reset the index could cause duplicate index issues during joins or filtering later in the workflow.
</p>

```python
import pandas as pd
import glob

# Get all CSV file paths
mubaweb_files = glob.glob("Data/raw_listings/mubaweb/*.csv")

# Read and combine them into one DataFrame
mubaweb_df = pd.concat(
    (pd.read_csv(file) for file in mubaweb_files),
    ignore_index=True
)

# Check the result
mubaweb_df.head()
```

<h2 style="color: #87BAC3;">🔤 Renaming Columns to English Standard for Mubawab</h2>

</p>

<h3>🔍 Technical Breakdown</h3>
<p>
<code>pd.DataFrame.rename(columns={...}, inplace=True)</code> applies a mapping dictionary that translates each source column name to its target English name. For example, <code>has_ascenceur</code> (French for elevator) becomes <code>has_elevator</code>, and <code>has_piscine</code> (French for pool) becomes <code>has_pool</code>. Note that both <code>has_ascenceur</code> and <code>has_ascenseur</code> (two common spelling variants in the scraped data) are mapped to the same target <code>has_elevator</code>, handling a real-world spelling inconsistency discovered during data exploration.
</p>

<h3>⚠️ Why This Matters</h3>
<p>
Schema harmonization across multi-source datasets is one of the most critical and error-prone steps in data integration. A single column name mismatch between two DataFrames will cause misaligned columns during concatenation, producing unexpected <code>NaN</code> values instead of properly merged data. By standardizing to English at the source level, the final merge becomes a straightforward union rather than a complex alignment exercise.
</p>

```python
# Get sets of columns
tayara_cols = set(tayara_df.columns)
mubaweb_cols = set(mubaweb_df.columns)

# All unique columns
all_cols = sorted(tayara_cols | mubaweb_cols)

# Prepare aligned DataFrame
comparison_list = []
for col in all_cols:
    tayara_val = col if col in tayara_cols else ""
    mubaweb_val = col if col in mubaweb_cols else ""
    comparison_list.append({"tayara_df": tayara_val, "mubaweb_df": mubaweb_val})

comparison_df = pd.DataFrame(comparison_list)
comparison_df
```

<h2 style="color: #87BAC3;">🧩 Injecting Missing Columns into Tayara from Mubawab Schema</h2>

<h3>🎯 Purpose</h3>
<p>
Based on the schema comparison performed in the previous step, this cell automatically adds all columns present in the Mubawab DataFrame but absent from the Tayara DataFrame, filling them with appropriate default values based on their data type. This is the final alignment step before the two datasets can be concatenated without schema conflicts.
</p>

<h3>🔍 Technical Breakdown</h3>
<p>
The code computes the set difference between Mubawab columns and Tayara columns, then iterates over each missing column. For each missing column, it checks the data type in the Mubawab DataFrame:
</p>
<ul>
<li><strong>Boolean columns</strong> are filled with <code>False</code> — assuming the absence of an amenity when no data is available is safer than assuming its presence.</li>
<li><strong>Numeric columns</strong> are filled with <code>np.nan</code> — a standard missing value representation for numerical features in Pandas and scikit-learn.</li>
<li><strong>All other columns</strong> (strings, objects, categories) are filled with <code>None</code> — the Python null equivalent, which Pandas stores as <code>NaN</code> for object dtypes.</li>
</ul>
<p>
The list of added columns is printed at the end for auditability.
</p>

<h3>⚠️ Why This Matters</h3>
<p>
Type-aware default value assignment is a sophisticated and important detail. Using <code>False</code> for boolean amenity flags reflects the real-world assumption that Tayara listings without an explicit elevator flag do not have an elevator — rather than leaving this as an unknown. Using <code>np.nan</code> for numerics preserves the distinction between "zero" and "unknown", which is critical for imputation strategies and model interpretability. This step ensures the concatenation operation produces a clean, type-consistent merged DataFrame rather than one polluted with mistyped defaults.
</p>

```python
import numpy as np

# Columns missing in tayara
missing_cols = set(mubaweb_df.columns) - set(tayara_df.columns)

for col in missing_cols:
    dtype = mubaweb_df[col].dtype

    # Boolean → set False
    if dtype == bool:
        tayara_df[col] = False

    # Numeric → use np.nan
    elif np.issubdtype(dtype, np.number):
        tayara_df[col] = np.nan

    # Everything else (object, string, category, etc.)
    else:
        tayara_df[col] = None

print("Added columns:", missing_cols)
```

<h2 style="color: #87BAC3;">🏗️ Loading, Standardizing, and Aligning BigDatis Listings</h2>

This comprehensive cell handles the complete standardization workflow for the <strong>BigDatis</strong> platform dataset in a single block. Following the same principles applied to Mubawab and Tayara, it loads the raw files, removes irrelevant columns, renames fields to the shared schema, adds missing columns, normalizes categorical values, and validates the result — all within one organized, commented pipeline.
</p>

<h3>🔍 Technical Breakdown</h3>
<p>
The cell is structured into seven sequential stages:
</p>
<ul>
<li><strong>Stage 1 — Load:</strong> All BigDatis CSV files are discovered and concatenated using the same <code>glob</code> + <code>pd.concat</code> pattern. The shape and head are printed for immediate validation.</li>

A mapping dictionary is applied via <code>.map(lambda x: transaction_map.get(x, x))</code>, which translates known values while safely passing through any unknown values unchanged.</li>
<li><strong>Stage 6 — Verify Shape and Columns:</strong> Shape and column list are printed to confirm the result.</li>
<li><strong>Stage 7 — Deduplication Check:</strong> Before the merge, all three DataFrames are checked for duplicate column names — a common issue when scraping introduces repeated header rows. Duplicate columns are removed using <code>.loc[:, ~df.columns.duplicated()]</code>.</li>
</ul>

<h3>⚠️ Why This Matters</h3>
<p>
The transaction type normalization step is particularly important. Mixing French and English labels in the same column after the merge would break any analysis or model that relies on <code>transaction_type</code> as a categorical filter. The duplicate column detection step addresses a subtle but dangerous issue: if a CSV file contains duplicate column headers (which can happen due to malformed scraping output), Pandas will silently keep both, leading to unpredictable behavior during merges and aggregations.

```python
# ─────────────────────────────────────────────
# 1. Load all bigdatis CSV files
# ─────────────────────────────────────────────
bigdatis_files = glob.glob("DATA/raw_listings/big_datis/*.csv")

bigdatis_df = pd.concat(
    (pd.read_csv(file) for file in bigdatis_files),
    ignore_index=True
)

print("Shape after loading:", bigdatis_df.shape)
print(bigdatis_df.head())

# ─────────────────────────────────────────────
# 2. Drop columns not needed for the merge
# ─────────────────────────────────────────────
bigdatis_df = bigdatis_df.drop(columns=["price_type"])
bigdatis_df = bigdatis_df.drop(columns=["district"])
bigdatis_df = bigdatis_df.drop(columns=["language"])

# ─────────────────────────────────────────────
# 3. Rename columns to match the shared schema
# ─────────────────────────────────────────────
bigdatis_df.rename(columns={
    "id"                  : "listing_id",
    "link"                : "listing_url",
    "surface"             : "surface_m2",
    "number_of_rooms"     : "rooms",
    "number_of_salle_de_bain": "bathrooms",
    "total_floor"         : "total_floors",
    "image_url"           : "image_urls",
    "scraped_at"          : "scrape_timestamp",
    "has_ascenseur"       : "has_elevator",
    "has_cave"            : "has_basement",
    "has_chauffage"       : "has_heating",
    "has_climatisseur"    : "has_air_conditioning",
    "has_jardin"          : "has_garden",
    "has_meubles"         : "has_furniture",
    "has_piscine"         : "has_pool",
    "has_terrasse"        : "has_terrace",
    "has_vue_sur_mer"     : "has_sea_view",
}, inplace=True)

# ─────────────────────────────────────────────
# 4. Add missing columns (present in mubawab/tayara but not in bigdatis)
# ─────────────────────────────────────────────
bigdatis_df["country"]      = "Tunisia"
bigdatis_df["total_rooms"]  = None
bigdatis_df["last_updated"] = None
bigdatis_df["has_standing"] = False

# floor already exists in bigdatis, keep it as-is

# ─────────────────────────────────────────────
# 5. Normalize transaction_type to match the shared schema
#    (bigdatis uses "Vente" / "Location", others use "Sale" / "Rent")
# ─────────────────────────────────────────────
transaction_map = {
    "Vente"    : "Sale",
    "Location" : "Rent",
}
bigdatis_df["transaction_type"] = bigdatis_df["transaction_type"].map(
    lambda x: transaction_map.get(x, x)
)

# ─────────────────────────────────────────────
# 6. Verify final columns
# ─────────────────────────────────────────────
print("\nShape after treatment:", bigdatis_df.shape)
print("\nColumns:", bigdatis_df.columns.tolist())

# ─────────────────────────────────────────────
# 7. (Optional) Compare columns across the three datasets
#    Uncomment once mubaweb_df and tayara_df are also loaded
# ─────────────────────────────────────────────


# ─────────────────────────────────────────────
# Check for duplicates in each df
print("mubaweb duplicates:", mubaweb_df.columns[mubaweb_df.columns.duplicated()].tolist())
print("tayara duplicates: ", tayara_df.columns[tayara_df.columns.duplicated()].tolist())
print("bigdatis duplicates:", bigdatis_df.columns[bigdatis_df.columns.duplicated()].tolist())

# Remove duplicate columns (keeps the first occurrence)
mubaweb_df  = mubaweb_df.loc[:, ~mubaweb_df.columns.duplicated()]
tayara_df   = tayara_df.loc[:, ~tayara_df.columns.duplicated()]
bigdatis_df = bigdatis_df.loc[:, ~bigdatis_df.columns.duplicated()]
```

<h2 style="color: #87BAC3;">🗺️ Loading, Standardizing, and Aligning Tunisie Annonce Listings</h2>

<h3>🎯 Purpose</h3>
<p>
This comprehensive cell applies the full standardization workflow to the <strong>Tunisie Annonce</strong> platform dataset. It introduces one unique challenge not present in the other datasets: the geographic information for this platform is stored in a single compound <code>region</code> string (e.g., "Localisation Tunisie>Tunis>La Marsa>El Aouina") that must be parsed and split into the shared <code>governorate</code> and <code>city</code> fields before the merge.
</p>

<h3>🔍 Technical Breakdown</h3>
<p>
The standardization is organized into seven stages:
</p>
<ul>
<li><strong>Stage 1 — Load:</strong> All Tunisie Annonce CSV files are loaded and concatenated. Shape and head are verified immediately.</li>
<li><strong>Stage 2 — Drop:</strong> <code>price_type</code> and <code>language</code> are removed. Note that <code>district</code> is not explicitly dropped here since the geographic parsing step (Stage 3) replaces the entire <code>region</code> field.</li>
<li><strong>Stage 3 — Parse Region:</strong> A custom <code>parse_region()</code> function is defined and applied using <code>.apply()</code>. This function splits the compound region string on the "&gt;" delimiter and extracts the governorate (index 1) and city (index 2) as separate fields. The index 0 ("Localisation Tunisie") and index 3 (district) are discarded. Null-safe handling is included via <code>pd.isna()</code> to prevent errors on missing values.</li>
<li><strong>Stage 4 — Rename:</strong> Platform-specific field names are mapped to the shared English schema, following the same renaming conventions applied to BigDatis.</li>
<li><strong>Stage 5 — Add Missing Columns:</strong> <code>country</code>, <code>total_rooms</code>, <code>last_updated</code>, <code>scrape_timestamp</code>, and <code>has_standing</code> are added with appropriate defaults.</li>
<li><strong>Stage 6 — Normalize transaction_type:</strong> French labels ("Vente", "Location") are mapped to English equivalents ("Sale", "Rent") using the same translation dictionary as BigDatis.</li>
<li><strong>Stage 7 — Verify:</strong> Shape, column list, and duplicate column check are printed.</li>
</ul>

<h3>⚠️ Why This Matters</h3>
<p>
The region parsing logic is the most technically sophisticated standardization step across all four platforms. It demonstrates an important principle: when geographic information is stored in compound strings rather than structured fields, a custom parsing function must be carefully designed to handle edge cases — including missing values, variable-length strings, and unexpected formats. The null-safe implementation using <code>pd.isna()</code> ensures that listings without a region string do not crash the pipeline. The discarding of the district-level component (index 3) is consistent with the decision made during Mubawab and BigDatis standardization to exclude district-level geography from the shared schema, maintaining geographic field consistency across all four datasets.
</p>

```python
import pandas as pd
import glob

# ─────────────────────────────────────────────
# 1. Load the tunisie_annonce CSV file(s)
#    (adjust the path/glob to match your folder structure)
# ─────────────────────────────────────────────
tunisie_files = glob.glob("DATA/raw_listings/tunisieannonce/*.csv")

tunisie_df = pd.concat(
    (pd.read_csv(file) for file in tunisie_files),
    ignore_index=True
)

print("Shape after loading:", tunisie_df.shape)
print(tunisie_df.head())

# ─────────────────────────────────────────────
# 2. Drop columns not needed for the merge
# ─────────────────────────────────────────────
tunisie_df = tunisie_df.drop(columns=["price_type"])
tunisie_df = tunisie_df.drop(columns=["language"])

# ─────────────────────────────────────────────
# 3. Parse 'region' into 'governorate' and 'city'
#    Raw format: "Localisation Tunisie>Tunis>La Marsa>El Aouina"
#    We take index [1] as governorate and [2] as city
# ─────────────────────────────────────────────
def parse_region(region_str):
    if pd.isna(region_str):
        return pd.Series({"governorate": None, "city": None})
    parts = region_str.split(">")
    # parts[0] = "Localisation Tunisie", parts[1] = governorate, parts[2] = city, parts[3] = district (ignored)
    governorate = parts[1].strip() if len(parts) > 1 else None
    city        = parts[2].strip() if len(parts) > 2 else None
    return pd.Series({"governorate": governorate, "city": city})

tunisie_df[["governorate", "city"]] = tunisie_df["region"].apply(parse_region)
tunisie_df = tunisie_df.drop(columns=["region"])

# ─────────────────────────────────────────────
# 4. Rename columns to match the shared schema
# ─────────────────────────────────────────────
tunisie_df.rename(columns={
    "id"                     : "listing_id",
    "link"                   : "listing_url",
    "surface"                : "surface_m2",
    "number_of_rooms"        : "rooms",
    "number_of_salle_de_bain": "bathrooms",
    "total_floor"            : "total_floors",
    "image_url"              : "image_urls",
    "has_ascenseur"          : "has_elevator",
    "has_cave"               : "has_basement",
    "has_chauffage"          : "has_heating",
    "has_climatisseur"       : "has_air_conditioning",
    "has_jardin"             : "has_garden",
    "has_meubles"            : "has_furniture",
    "has_piscine"            : "has_pool",
    "has_terrasse"           : "has_terrace",
    "has_vue_sur_mer"        : "has_sea_view",
}, inplace=True)

# ─────────────────────────────────────────────
# 5. Add missing columns (present in other dfs but not here)
# ─────────────────────────────────────────────
tunisie_df["country"]          = "Tunisia"
tunisie_df["total_rooms"]      = None
tunisie_df["last_updated"]     = None
tunisie_df["scrape_timestamp"] = None
tunisie_df["has_standing"]     = False

# ─────────────────────────────────────────────
# 6. Normalize transaction_type
#    tunisie_annonce uses "Vente" / "Location" → "Sale" / "Rent"
# ─────────────────────────────────────────────
transaction_map = {
    "Vente"    : "Sale",
    "Location" : "Rent",
}
tunisie_df["transaction_type"] = tunisie_df["transaction_type"].map(
    lambda x: transaction_map.get(x, x)
)

# ─────────────────────────────────────────────
# 7. Verify
# ─────────────────────────────────────────────
print("\nShape after treatment:", tunisie_df.shape)
print("\nColumns:", tunisie_df.columns.tolist())
print("\nDuplicate columns:", tunisie_df.columns[tunisie_df.columns.duplicated()].tolist())
```

<h2 style="color: #87BAC3;">📥 Loading Houni Blog Articles</h2>

<h3>🎯 Purpose</h3>
<p>
This cell loads the blog articles scraped from <strong>Houni.tn</strong>, a Tunisian real estate information platform that publishes neighborhood guides, mortgage comparison articles, legal advice, and lifestyle-oriented property content. Houni's editorial tone is more consumer-facing and educational, targeting first-time buyers and renters rather than professional investors.
</p>

<h3>🔍 Technical Breakdown</h3>
<p>
The Houni CSV was scraped with a character encoding inconsistency: the file was saved in <strong>Latin-1</strong> (ISO-8859-1) rather than UTF-8, causing French accented characters (é, è, ê, à, ç, etc.) to appear as garbled sequences (e.g., <code>crÃ©dit</code> instead of <code>crédit</code>) when read naively. To fix this, the file is first loaded with <code>encoding="latin-1"</code> to prevent parsing errors, and then each string value is re-encoded to Latin-1 bytes and decoded as UTF-8 — the standard mojibake correction technique. The <code>fix_mojibake()</code> helper function handles this safely, with a try/except block that leaves values unchanged if the correction fails (e.g., for already-clean strings or irreparable sequences).
</p>

<h3>⚠️ Why This Matters</h3>
<p>
Encoding corruption is one of the most common and damaging issues in web-scraped text data. If left uncorrected, garbled characters would propagate into every downstream NLP operation — breaking tokenizers, polluting vocabulary dictionaries, and producing nonsensical embeddings. Fixing the encoding at the loading stage is the only correct remedy: it ensures that the Houni articles enter the pipeline as clean, human-readable French text, fully compatible with French-language NLP models and tokenizers.
</p>

```python
# Load Houni blog articles — requires encoding fix (Latin-1 mojibake)
def fix_mojibake(text):
    """Fix text incorrectly read as Latin-1 that was originally UTF-8."""
    if not isinstance(text, str):
        return text
    try:
        return text.encode('latin-1').decode('utf-8')
    except (UnicodeDecodeError, UnicodeEncodeError):
        return text  # Return as-is if correction fails

houni_blog_df = pd.read_csv("Data/articles/houni/houni_blog_articles.csv", encoding="latin-1")
houni_blog_df = houni_blog_df.apply(lambda col: col.map(fix_mojibake))

print("Shape:", houni_blog_df.shape)
print("Columns:", houni_blog_df.columns.tolist())
houni_blog_df.head()
```

```python
# Vocabulary richness by source
def vocab_stats(text_series):
    all_words = []
    for t in text_series.fillna(""):
        all_words.extend(re.findall(r'\b[a-zéèêëàâùûüôîïç]{3,}\b', t.lower()))
    total = len(all_words)
    unique = len(set(all_words))
    return pd.Series({"total_words": total, "unique_words": unique,
                      "ttr": round(unique / total, 4) if total > 0 else 0})

vocab_df = art.groupby("source")["content"].apply(vocab_stats).reset_index()
vocab_df.columns = ["source", "metric", "value"]
vocab_pivot = vocab_df.pivot(index="source", columns="metric", values="value")
print("Vocabulary richness by source:")
print(vocab_pivot.to_string())

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
metrics = ["total_words", "unique_words", "ttr"]
titles  = ["Total Words", "Unique Words", "Type-Token Ratio (TTR)"]
for ax, metric, title in zip(axes, metrics, titles):
    vals = vocab_pivot[metric]
    ax.bar(vals.index, vals.values, color=PALETTE[:len(vals)])
    ax.set_title(title)
    ax.set_ylabel(metric)
    for i, v in enumerate(vals.values):
        ax.text(i, v * 1.01, f"{v:,.3f}" if metric == "ttr" else f"{v:,.0f}",
                ha="center", fontsize=9)

plt.tight_layout()
plt.savefig("Data/eda_vocab_richness.png", bbox_inches="tight")
plt.show()
```
